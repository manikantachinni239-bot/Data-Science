```python
def detect_loop_LL(start):
    slow=fast=start
    while fast!=None and fast["link"]!=None:
        slow=slow["link"]
        fast=fast["link"]["link"]
        if slow is fast:
            return True
            return False
start={"data":2,"link":None}
node2={"data":3,"link":None}
node3={"data":4,"link":None}
node4={"data":5,"link":None}
node5={"data":6,"link":None}
start["link"]=node2
node2["link"]=node3
node3["link"]=node4
node4["link"]=node5
node5["link"]=node4
print(detect_loop_LL(start))
```

```python
def reverse_LL(start):
    current=start
    previous=None
    while current!=None:
        tmp=current["link"]
        current["link"]=previous
        previous=current
        current=tmp
    return previous
start={'data':1,'link':{'data':2,
       'link':{'data':3,
               'link':{'data':4,
                       'link':{'data':5,
                               'link':{'data':6,
                                       'link':{'data':7,'link':{'data':8,'link':None}}}}}}}}
print(reverse_LL(start))        
```

```python
def matrix_addition(matrix1,matrix2):
    result=[[0,0],[0,0]]
    for i in range(len(matrix1)):
        for j in range(len(matrix2)):
            result[i][j]=matrix1[i][j]+matrix2[i][j]
    return result
matrix1=[[1,2],[3,4]]
matrix2=[[5,6],[7,8]]
print(matrix_addition(matrix1,matrix2))        
```

```python
def trace_matrix(matrix):
    trace=0
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            if i==j:
                trace+=matrix[i][j]
        return trace
matrix=[[1,2],[4,5],[7,8]]
print(trace_matrix(matrix))
```

```python
def searching(matrix,key):
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if matrix[i][j]==key:
              return True
        return False
matrix=[[1,2,3],[4,5,6],[7,8,9]]
print(searching(matrix[1]))                
```

```python
def transpose_matrix(matrix):
    for i in range(0,len(matrix[0])):
        for j in range(0,len(matrix[0])):
            if i!=j:
                matrix[i][j],matrix[j][i]=matrix[i][j],matrix[j][i]
    return matrix             
```

```python
def check_identity_matrix(matrix):
    for i in range(0,len(matrix)):
        for j in range(len(matrix[i])):
            if i==j:
                if matrix[i][j]!=1:
                    return False
            else:
                if matrix[i][j]!=0:
                    return False
    return True
matrix=[[1,0,1],[0,1,0],[0,0,1]]
check_identity_matrix(matrix)           
```

```python
def stimulate_CLT(population,no_of_samples,sample_size):
    
    no_of_samples=1000
    sample_size=30
    samples_mean=[]
    for i in range(0,no_of_samples):
        samples=np.random.choice(population,size=sample_size)
        samples_mean.append(np.mean(samples))
    return np.mean(samples_mean)    
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import pareto,boxcox
```

```python
def bootstraping(x,ci,sample_size,no_of_sample):
    lst=[]
    for i in range(no_of_sample):
        sample = np.random.choice(x,sample_size)
        median = np.median(sample)
        lst.append(median)
    lst.sort()
    n=int(no_of_sample*(ci/100))
    return (lst[0],lst[n])
```

```python
x=[np.random.randint(50,80) for i in range(100)]
ci=95
sample_size=30
no_of_sample=1000
bootstraping(x,ci,sample_size,no_of_sample)
```

```python
def stimulate_boot(population,x):
    sum_population=1000
    samples=30
    res=[]
    for i in range(sum_population):
        choice=np.random.choice(population,samples)
        ss=np.median(choice)
        res.append(ss)
    res.sort()
    return np.percentile(res,x)
val=np.random.randint(1,100,size=100)
result=stimulate_boot(val,95)
print(result)
```

```python
def permutation_test(sample1,sample2,no_samples):
    original_delta=np.mean(sample1)-np.mean(sample2)
    jumbled_sample=np.concatenate([sample1,sample2])
    samples_diff=[]
    for i in range(no_samples):
        np.random.shuffle(jumbled_sample)
        x=jumbled_sample[:len(sample1)]
        y=jumbled_sample[len(sample1):len(sample1)+len(sample2)]
        diff=np.mean(x)-np.mean(y)
        samples_diff.append(diff)
    samples_diff=np.array(samples_diff)
    sorted_sample_diff=np.sort(samples_diff)
    p_value=(np.sum(sorted_sample_diff>original_delta)/no_samples)*100
    return p_value
```

```python
sample1=np.random.randn(50)
sample2=np.random.randn(50)
no_samples=1000
permutation_test(sample1,sample2,no_samples)
```

```python
# LL------->lower limit,UL---------->upper limit.
def detecting_outliers_with_IQR(column,factor):
    IQR=np.quantile(column,0.75)-np.quantile(column,0.25)
    LL=np.quantile(column,0.25)-(factor*IQR)
    UL=np.quantile(column,0.75)+(factor*IQR)
    return LL,UL
```

```python
def jobSequencing(deadline,profit):
    n=len(profit)
    jobs=[]
    for i in range(n):
        jobs.append((profit[i],deadline[i]))
    jobs.sort(reverse=True)
    max_deadline=max(deadline)
    slots=[False]*max_deadline
    job_count=0
    total_profit=0
    for p,d in jobs:
        for j in range(d-1,-1,-1):
            if not slots[j]:
                slots[j]=True
                job_count += 1
                total_profit+= p
                break
    return [job_count,total_profit]
profit=[20, 10, 40, 30]
deadline=[4, 1, 1, 1]
jobSequencing(deadline,profit)
```

```python
def minimum_no_platform(arrival,departure):
    lst=[]
    count=1
    for i in range(0,len(departure)):
        lst.append((arrival[i],departure[i]))
    lst.sort(key=lambda x:x[0],reverse=True)
    for j in range(0,len(lst)-1):
        if lst[j+1][0]<=lst[j][1]:
            count+=1
    return count
arrival=[1000,935,1100]
departure=[1200,1240,1130]
minimum_no_platform(arrival,departure)    
```

```python
class graph:
    def __init__(self):
        self.graph={}
    def create_node(self,vertex):
        if vertex not in self.graph:
            self.graph[vertex]=[]
    def add_edge(self,vertex1,vertex2):
        if vertex1 not in self.graph:
            self.create_node(vertex1)
        if vertex2 not in self.graph:
            self.create_node(vertex2)
        self.graph[vertex1].append(vertex2)
        self.graph[vertex2].append(vertex1)
    def display_graph(self):
        print(self.graph)
```

```python
def dfs(graph,start):
    stack=[]
    stack.append(start)
    dfs_visited=set()
    while stack:
        if stack[-1] not in dfs_visited:
            dfs_visited.add(stack[-1])
        element=stack.pop()
        for i in graph[element]:
            if i not in dfs_visited:
                stack.append(i)
    return  dfs_visited
graph={'A': ['B'], 'B': ['A', 'C'], 'C': ['B'], 'D': ['E'], 'E': ['D', 'F'], 'F': ['E']}
dfs(graph,"A")
```

```python
# check the graqph is connected or not.
def graph_connected_or_not(graph,start):
    var=dfs(graph,start)
    return len(graph)==len(var)
graph={'A': ['B'], 'B': ['A', 'C'], 'C': ['B'], 'D': ['E'], 'E': ['D', 'F'], 'F': ['E']}
graph_connected_or_not(graph,"A")
```

```python
def bfs(graph,start):
    queue=[]
    queue.append(start)
    bfs_visited=set()
    while queue:
        if queue[0] not in bfs_visited:
            bfs_visited.add(queue[0])
        element=queue.pop(0)
        for i in graph[element]:
            if i not in bfs_visited:
                queue.append(i)
    return  bfs_visited
graph={'A': ['B'], 'B': ['A', 'C'], 'C': ['B'], 'D': ['E'], 'E': ['D', 'F'], 'F': ['E']}
bfs(graph,"E")
```

```python
# check a graph contains issolated vertex(no edge is coming to it) or not.
def check_isolated_vertex(graph):
    if len(graph)==0:
        return "Graph is empty"
    for a,b in graph.items():
        if b==[]:
            return True
    return False
graph={'A': ['B'], 'B': ['A', 'C'], 'C': ['B'], 'D': ['E'], 'E': ['D', 'F'], 'F': ['E']}
check_isolated_vertex(graph)
```

```python
# Eular graph means the edges contains even degrees.
def check_graph_eular_graph_or_not(graph):
    for i in graph:
        if len(graph[i])%2==0:
            return "Even"
        else:
            return "Odd"
graph={'A': ['B'], 'B': ['A', 'C'], 'C': ['B'], 'D': ['E'], 'E': ['D', 'F'], 'F': ['E']}
check_graph_eular_graph_or_not(graph)
```

```python
def detect_cycle(graph,start):
    for key,value in graph.items():
        if start in value:
            return True
    return False
graph={'A': ['B'], 'B': ['A', 'C'], 'C': ['B']}
detect_cycle(graph,"A")   
```
